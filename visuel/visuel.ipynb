```python
import pygame
import sys

# Initialize Pygame
pygame.init()

# Set up the window
screen_width, screen_height = 800, 600
screen = pygame.display.set_mode((screen_width, screen_height))
pygame.display.set_caption('Planning Poker Game')

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREY = (200, 200, 200)
GREEN = (0, 200, 0)

# Define button properties
button_width, button_height = 200, 50
button_start_x = screen_width // 2 - button_width // 2
button_start_y = 150
button_gap = 70

# Game states
main_menu = True
rules_menu = False
load_menu = False

# State variables for rule menu buttons
strictes_clicked = True
moyenne_clicked = False
mediane_clicked = False

# Function definitions

def draw_button(text, y, active_color, inactive_color, clicked=False):
    mouse = pygame.mouse.get_pos()
    click = pygame.mouse.get_pressed()

    if clicked:
        pygame.draw.rect(screen, active_color, [button_start_x, y, button_width, button_height])
    elif button_start_x < mouse[0] < button_start_x + button_width and y < mouse[1] < y + button_height:
        pygame.draw.rect(screen, active_color, [button_start_x, y, button_width, button_height])
        if click[0] == 1:
            return True
    else:
        pygame.draw.rect(screen, inactive_color, [button_start_x, y, button_width, button_height])

    small_text = pygame.font.Font("freesansbold.ttf", 20)
    text_surf, text_rect = text_objects(text, small_text)
    text_rect.center = ((button_start_x + (button_width // 2)), (y + (button_height // 2)))
    screen.blit(text_surf, text_rect)
    return False



# Main game loop
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False 

    screen.fill(WHITE)

    if main_menu:
        if draw_button("Start", button_start_y, GREY, GREEN):
            main_menu = False
            load_menu = True
        if draw_button("Rules", button_start_y + button_gap, GREY, GREEN):
            main_menu = False
            rules_menu = True
        if draw_button("Quit", button_start_y + 2 * button_gap, GREY, GREEN):
            running = False
    
    if load_menu:
        if draw_button("Load Game", button_start_y, GREY, GREEN):
            print("Load Game clicked")
        if draw_button("Players", button_start_y + button_gap, GREY, GREEN):
            print("Players clicked")
        if draw_button("Start", button_start_y + 2 * button_gap, GREY, GREEN):
            print("Start clicked")
        if draw_button("Back to Main Menu", button_start_y + 3 * button_gap, GREY, GREEN):
            # Go back to main menu
            load_menu = False
            main_menu = True

    if rules_menu:
        base_color = GREEN
        base_hover_color = GREY

        if draw_button("Strictes", button_start_y, base_hover_color, base_color, strictes_clicked):
            strictes_clicked = True
            moyenne_clicked = False  # Reset the other buttons
            mediane_clicked = False
        elif draw_button("Moyenne", button_start_y +  button_gap, base_hover_color, base_color, moyenne_clicked):
            moyenne_clicked = True
            strictes_clicked = False  # Reset the other buttons
            mediane_clicked = False
        elif draw_button("Médiane", button_start_y + 2 * button_gap, base_hover_color, base_color, mediane_clicked):
            mediane_clicked = True
            strictes_clicked = False  # Reset the other buttons
            moyenne_clicked = False

        if draw_button("Back to Main Menu", button_start_y + 3 * button_gap, base_hover_color, base_color):
            rules_menu = False
            main_menu = True
            
    pygame.display.flip()

pygame.quit()
sys.exit()
```
